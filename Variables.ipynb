```python
# why use variables?
# Variables allow us to store and manipulate data in a program.
# different types of data can be stored in variables, such as numbers, text, lists, and etc.
# By using variables, we can easily reference and update values throughout our code without having to rewrite the actual data.
# Variables also improve code readability and maintainability by giving meaningful names to data.
# Example of using variables in Python:

# Syntax => [Variable Name] [Assignment Operator] [Value]

name = "Alber Gamal"
age = 22
faculty = "Artificial Intelligence"

print("Hello, my name is", name, "and I am", age, "years old.")
print("I study at the faculty of", faculty, ".")
```

```python
# Name of Variable Rules:
# 1. Variable names must start with a letter (a-z, A-Z) or an underscore (_).
# 2. Cannot start with a number (0-9) or special characters (e.g., @, #, $, %, etc.).
# 3. Can include (0-9), or underscore.
# 4. Cannot include spaces, or special characters. 
# 5. Cannot use reserved keywords (e.g., if, else, while, for, etc.) as variable names.
# 6. Variable names are case-sensitive (e.g., age and Age are different variables).

# Example of valid variable names

first_name = "Alber"
_lastName = "Gamal"
age2 = 30
address_line_1 = "Egypt, Alexandria"

# Example of invalid variable names

# 1stName = "Jane"  # starts with a number
# user-name = "johndoe"  # contains a special character
# my age = 25  # contains a space
# if = 10  # uses a reserved keyword

# To see the list of reserved keywords in Python, you can use the following code:
print(help("keywords"))

# Following these rules helps ensure that your code is syntactically correct and easy to understand.
```

```python
# Variable Naming Conventions
# 1. Normal Case: Used for single-word variable names.
# 2. camelCase: Used for multi-word variable names where the first word is lowercase and subsequent words are capitalized.
# 3. snake_case: Used for multi-word variable names where words are separated by underscores

# Examples:
name = "Alber Gamal" # Single Word => Normal
myName = "Alber Gamal" # Two Words => camelCase
my_name = "Alber_Gamal" # Two Words => snake_case
```
